```python
# Inserindo as colunas em uma lista
cols = ['id_pedido', 'id_consumidor', 'status_da_ordem', 'horario_pedido',
       'pedido_aprovado', 'pedido_com_entregador', 'pedido_entregue',
       'data_estimada', 'item_id', 'produto_id', 'vendedor_id',
       'data_limite_envio', 'preco', 'frete', 'pagamento_sequencial',
       'tipo_pagamento', 'prestacoes', 'valor_pagamento', 'id_review',
       'review_score', 'titulo_review', 'comentario_review',
       'data_criacao_review', 'data_resposta_review', 'categoria_produto',
       'qtd_fotos_produto', 'peso_produto', 'comprimento_produto',
       'altura_produto', 'largura_produto', 'id_unico_consumidor',
       'cep_consumidor', 'cidade_consumidor', 'estado_consumidor',
       'cep_vendedor', 'cidade_vendedor', 'estado_vendedor']
```

```python
# Selecionando a variável tipo de pagamento e categorizando-o para o modelo de ML
df = pd.read_csv("BaseDadosLimpa.csv", usecols=cols, dtype={'tipo_pagamento': 'category'})
```

```python
df.info()
```

```python
# Pedindo para o pandas códigos correspondentes aos tipos de pagamento (cartão de crédito, boleto, débito...)
df['tps_pagamentos'] = df['tipo_pagamento'].cat.codes
```

```python
# Exibindo os tipos de pagamentos e o número correspondente de entradas.
df['tps_pagamentos'].value_counts()
```

```python
df.head()
```

```python
# Verificando quando foi a última compra do cliente
df_compra = df.groupby('id_unico_consumidor').horario_pedido.max().reset_index()
df_compra.columns = ['id_unico_consumidor', 'DataMaximaCompra']
df_compra['DataMaximaCompra'] = pd.to_datetime(df_compra['DataMaximaCompra'])
df_compra.head()
```

```python
# Seleção do ponto de observação com a data máxima de compra
df_compra['Recencia'] = (df_compra['DataMaximaCompra'].max() - df_compra['DataMaximaCompra']).dt.days
df_compra.head()
```

```python
# Juntando essa informação no banco a partir do id único do consumidor
df_usuario = pd.merge(df, df_compra[['id_unico_consumidor', 'Recencia']], on='id_unico_consumidor')
df_usuario.head()
```

```python
# Extraindo a freqüência de compras, ou seja, quantos pedidos um cliente realizou.
df_frequencia = df.groupby('id_unico_consumidor').pedido_aprovado.count().reset_index()
df_frequencia.columns = ['id_unico_consumidor', 'Frequencia']
df_frequencia.head()
```

```python
# Adicionando também os dados da frequência no banco usuário
df_usuario = pd.merge(df_usuario, df_frequencia, on='id_unico_consumidor')
df_usuario.head()
```

```python
# Extraindo a receita
df_receita = df.groupby('id_unico_consumidor').valor_pagamento.sum().reset_index()
df_receita.columns = ['id_unico_consumidor', 'Receita']
df_receita.head()
```

```python
df_usuario = pd.merge(df_usuario, df_receita, on='id_unico_consumidor')
df_usuario.head()
```
